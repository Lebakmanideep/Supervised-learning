```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("wbcd.csv")
```

```python
df.head()
```

```python
# replace the B and M in column 2 as Benign and Malignant

df.diagnosis = np.where(df.diagnosis == "B", 'Benign', df.diagnosis)
df.diagnosis= np.where(df.diagnosis == "M", 'Malignant', df.diagnosis)
```

```python
df.head()
```

```python
# removeing the column ID

df = df.iloc[:,1:]
df
```

```python
# let's do the scaling 
# Normalization
def norm(i):
    x = (i - i.min())/(i.max() - i.min())
    return x
```

```python
df_n = norm(df.iloc[:,1:])
```

```python
df_n.describe()
```

```python
x = np.array(df_n)
y = np.array(df.diagnosis)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 13)
```

```python
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

```python
pd.crosstab(y_test, pred, rownames = ['Actual'], colnames = ['Predictions'])
```

```python
pred_train = knn.predict(x_train)
pred_train
```

```python
accuracy_score(y_train, pred_train)
```

```python
pd.crosstab(y_train, pred_train, rownames = ['actual'], colnames = ['predictions'])
```

```python
# creating a list
acc = []
```

```python
for i in range(3,50,2):
    neigh = KNeighborsClassifier(n_neighbors = i)
    neigh.fit(x_train, y_train)
    train_acc = np.mean(neigh.predict(x_train)==y_train)
    test_acc = np.mean(neigh.predict(x_test)==y_test)
    acc.append([train_acc,test_acc])
```

```python
len(acc)
```

```python
# lets plot this in a graph

import matplotlib.pyplot as plt
```

```python
plt.plot(np.arange(3,50,2),[i[0] for i in acc],'ro-')
plt.plot(np.arange(3,50,2),[i[1] for i in acc],'bo-')
```
